````python
# Convert Start_Time and End_Time to datetime and calculate Impact_Duration_HR
def hours_minutes_decimal(timedelta_series):
    """
    Convert timedelta to hours.minutes format (e.g., 1.14 for 1 hour 14 minutes)
    """
    total_seconds = timedelta_series.dt.total_seconds()
    hours = total_seconds // 3600
    minutes = (total_seconds % 3600) // 60
    return hours + (minutes / 100)  # This gives you the format you want


# Helper function to convert to datetime
def to_datetime(datetime_str):
    """
    Converts a string to a datetime object.
    """
    return_value = pd.NaT
    if datetime_str:
        if "." in datetime_str:
            before_dot, _ = datetime_str.rsplit(".", 1)
        else:
            before_dot = datetime_str
        return_value = datetime.strptime(before_dot, "%Y-%m-%d %H:%M:%S")
    return return_value


# Convert the Start_Time and End_Time to datetime
df["Start_Time_TS"] = df["Start_Time"].apply(to_datetime)
df["Start_Year"] = df["Start_Time_TS"].dt.year
df["End_Time_TS"] = df["End_Time"].apply(to_datetime)
df["Impact_Duration_HR"] = hours_minutes_decimal(
    df["End_Time_TS"] - df["Start_Time_TS"]
)
df = df.drop(columns=["Start_Time", "End_Time"], errors="ignore")
df.info()
````

````python

def plot_incidents(df:pd.DataFrame, 
                   group_by_col:str, 
                   top_n:int=10, 
                   date_col:str='Start_Time_TS', 
                   severity_col:str='Severity'):
    """
    Generates a generic bar chart of incident counts and a line chart of
    average severity, with an option to display only the top N groups.

    Args:
        df (pd.DataFrame): DataFrame with incident data.
        group_by_col (str): The column to group by ('month', 'week', 'weekday', etc.).
        top_n (int, optional): If provided, displays only the top N groups based on
                               incident count. Defaults to None (all groups).
        date_col (str): The name of the datetime column.
        severity_col (str): The name of the severity column.
    """
    df_copy = df.copy()
    if group_by_col == "Total_Vehicles" or group_by_col == "Total_Drivers":
       df_copy[group_by_col] = round(df_copy[group_by_col]/ 10**6, 2)


    # --- Feature Engineering for Time-Based Grouping ---
    grouping_cols = [group_by_col]
    if group_by_col == "month":
        df_copy[group_by_col] = df_copy[date_col].dt.month_name()
    elif group_by_col == "weekday":
        df_copy[group_by_col] = df_copy[date_col].dt.day_name()
    elif group_by_col == "Weather_Condition":
        grouping_cols = [group_by_col, 'Weather_Condition_Encoded']

    # --- Aggregation ---
    aggregated_data = df_copy.groupby(grouping_cols).agg(
        incident_count=(severity_col, 'size'),
        average_severity=(severity_col, 'mean')
    ).reset_index()

    # --- Filter for Top N if specified ---
    aggregated_data = aggregated_data.sort_values('incident_count', ascending=False).head(top_n)

    # --- Sorting for Plotting ---
    if group_by_col == 'weekday':
        weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
        aggregated_data[group_by_col] = pd.Categorical(aggregated_data[group_by_col], categories=weekday_order, ordered=True)
    elif group_by_col == "month":
        month_order = [
            "January",
            "February",
            "March",
            "April",
            "May",
            "June",
            "July",
            "August",
            "September",
            "October",
            "November",
            "December",
        ]
        aggregated_data[group_by_col] = pd.Categorical(
            aggregated_data[group_by_col], categories=month_order, ordered=True
        )

    aggregated_data[group_by_col] = (
        aggregated_data[grouping_cols].astype(str).agg(", ".join, axis=1)
    )
    aggregated_data = aggregated_data.sort_values(grouping_cols)

    # --- Plotting ---
    fig, ax1 = plt.subplots(figsize=(12, 6))
    xlabel = group_by_col.replace('_', ' ').title()
    if group_by_col == "Total_Vehicles" or group_by_col == "Total_Drivers":
        xlabel = f"{xlabel} (in Millions) - "
    xlabel = f"{xlabel} (Across California)"
    title = f'Top {top_n} {xlabel} by Incident Count'

    ax1.bar(aggregated_data[group_by_col], aggregated_data['incident_count'], color='cornflowerblue', label='Incident Count')
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel('Number of Incidents', color='cornflowerblue')
    ax1.tick_params(axis='y', labelcolor='cornflowerblue')
    ax1.tick_params(axis='x', rotation=45)
    ax2 = ax1.twinx()
    ax2.plot(aggregated_data[group_by_col], aggregated_data['average_severity'], color='red', marker='o', label='Average Severity')
    ax2.set_ylabel('Average Severity', color='red')
    ax2.tick_params(axis='y', labelcolor='red')

    plt.title(title)
    fig.tight_layout()
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    group_by_col = re.sub(r"\(.*?\)", "", group_by_col)
    plt.savefig(f'./images/incidents_by_{group_by_col}_{ANALYSIS_STATE_CODE}.png')
    plt.show()
````
